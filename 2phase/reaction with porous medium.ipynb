## Теория

<hr>

### Химия

Рассмотрим реакцию соляной кислоты с известняком (кальцитом):

$$
\begin{gathered}
\text{CaCO}_3 + 2 \text{HCl} \rightarrow \text{CaCl}_2 + \text{H}_2\text{O} + \text{CO}_2
\\
1 \hspace{1.18cm} 2 \hspace{1.15cm} 3 \hspace{1.12cm} 4 \hspace{0.9cm} 5
\end{gathered}
$$

В массовом виде относительно кальцита уравнение реакции выглядит так:
$$
\begin{gathered}
\text{CaCO}_3 + 0.729 \, \text{HCl} \rightarrow 1.109 \, \text{CaCl}_2 + 0.18 \, \text{H}_2\text{O} + 0.44 \, \text{CO}_2
\end{gathered}
$$

Массовые стехиометрические коэффициенты:

- $\chi_2 = 0.728579$
- $\chi_3 = 1.108884$
- $\chi_4 = 0.179992$
- $\chi_5 = 0.439703$

Предполагается, что смачивающая фаза, в основном состоящая из воды, несет в себе кислоту, полностью растворимые соль и углекислый газ

Кинетика реакции описывается формулой:

$$
J = k_a S (1 - \phi) \frac{M_1}{2} S_s \rho_2^m \frac{(c_2 - c_{\text{eq}})^m}{M_2^m}
$$

- $M_1 = 0.100088$ кг/моль
- $M_2 = 0.036461$ кг/моль

<hr>

### МСС

#### Уравнения неразрывности

$$
\frac{\partial \rho_A}{\partial t} + \nabla \cdot (\rho_A \vec{v}_A) = \chi_A J
$$

В одномерном случае получим систему уравнений

- для породы:

$$
\frac{\partial}{\partial t} \left[ \rho_1 (1 -\phi) \right] = - J
$$

- для воды и растворенных компонентов

$$
\begin{gathered}
\frac{\partial}{\partial t} \left[ \rho_2 \phi S c_2 \right] +
\frac{\partial}{\partial x} \left[ \rho_2 c_2 \phi S v_w \right] =
- \chi_2 J
\\
\frac{\partial}{\partial t} \left[ \rho_3 \phi S c_3 \right] +
\frac{\partial}{\partial x} \left[ \rho_3 c_3 \phi S v_w \right] =
\chi_3 J
\\
\frac{\partial}{\partial t} \left[ \rho_4 \phi S (1 - c_2 - c_3 - c_5) \right] +
\frac{\partial}{\partial x} \left[ \rho_4 (1 - c_2 - c_3 - c_5) \phi S v_w \right] =
\chi_4 J
\\
\frac{\partial}{\partial t} \left[ \rho_5 \phi S c_5 \right] +
\frac{\partial}{\partial x} \left[ \rho_5 c_5 \phi S v_w \right] =
\chi_5 J
\end{gathered}
$$

- для нефти:

$$
\frac{\partial}{\partial t} \left[ \rho_o \phi (1 - S) \right] + 
\frac{\partial}{\partial x} \left[ \rho_o \phi (1 - S) v_o \right] = 0
$$

Скорость фильтрации по определению:
$$
W_w = \phi S v_w
\hspace{1cm}
W_o = \phi (1 - S) v_o
$$

Из з. Дарси скорости фильтрации:

$$
\begin{gathered}
W_w = -k \frac{k_w}{\mu_w} \frac{\partial P}{\partial x}
\hspace{1cm}
W_o = -k \frac{k_o}{\mu_o} \frac{\partial P}{\partial x}
\\
W_o = W_w \frac{k_o}{k_w} \frac{\mu_w}{\mu_o}
\\
W_o + W_w = W_w \left( 1 + \frac{k_o}{k_w} \frac{\mu_w}{\mu_o} \right)
\\
\frac{W_w}{W_o + W_w} = \frac{1}{1 + \frac{k_o}{k_w} \frac{\mu_w}{\mu_o}} = b(S)
\\
W_w = W b
\hspace{1cm}
W_o = W ( 1 - b)
\end{gathered}
$$

Здесь пренебрегаются капиллярное давление, влияние кол-ва растворенного $\text{CO}_2$ в воде на $\mu_w$

Таким образом:
- для породы:
  
$$
\frac{\partial}{\partial t} \left[ \rho_1 (1 -\phi) \right] = - J
$$

- для воды и растворенных компонентов
  
$$
\begin{gathered}
\frac{\partial}{\partial t} \left[ \rho_2 \phi S c_2 \right] +
\frac{\partial}{\partial x} \left[ \rho_2 c_2 Wb \right] =
- \chi_2 J
\\
\frac{\partial}{\partial t} \left[ \rho_3 \phi S c_3 \right] +
\frac{\partial}{\partial x} \left[ \rho_3 c_3 Wb \right] =
\chi_3 J
\\
\frac{\partial}{\partial t} \left[ \rho_4 \phi S (1 - c_2 - c_3 - c_5) \right] +
\frac{\partial}{\partial x} \left[ \rho_4 (1 - c_2 - c_3 - c_5) Wb \right] =
\chi_4 J
\\
\frac{\partial}{\partial t} \left[ \rho_5 \phi S c_5 \right] +
\frac{\partial}{\partial x} \left[ \rho_5 c_5 Wb \right] =
\chi_5 J
\end{gathered}
$$

- для нефти:

$$
\frac{\partial}{\partial t} \left[ \rho_o \phi (1 - S) \right] + 
\frac{\partial}{\partial x} \left[ \rho_o W (1-b) \right] = 0
$$

Пусть жидкости и порода несжимаемы, тогда:

- для породы:
  
$$
\frac{\partial \phi}{\partial t} = \frac{1}{\rho_1} J
$$

- для воды и растворенных компонентов

$$
\begin{gathered}
\frac{\partial}{\partial t} \left[ \phi S c_2 \right] +
\frac{\partial}{\partial x} \left[  c_2 Wb \right] =
- \frac{\chi_2}{\rho_2} J
\\
\frac{\partial}{\partial t} \left[ \phi S c_3 \right] +
\frac{\partial}{\partial x} \left[ c_3 Wb \right] =
\frac{\chi_3}{\rho_3} J
\\
\frac{\partial}{\partial t} \left[ \phi S (1 - c_2 - c_3 - c_5) \right] +
\frac{\partial}{\partial x} \left[ (1 - c_2 - c_3 - c_5) Wb \right] =
\frac{\chi_4}{\rho_4} J
\\
\frac{\partial}{\partial t} \left[ \phi S c_5 \right] +
\frac{\partial}{\partial x} \left[ c_5 Wb \right] =
\frac{\chi_5}{\rho_5} J
\end{gathered}
$$

- для нефти:

$$
\frac{\partial}{\partial t} \left[ \phi (1 - S) \right] + 
\frac{\partial}{\partial x} \left[ W (1-b) \right] = 0
$$

Сложив уравнения для воды и компонентов, получим:
$$
\frac{\partial \phi S}{\partial t} +
\frac{\partial Wb}{\partial x} =
J \left( -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
$$

Складывая это уравнение с уравнением неразрывности для нефти получим:
$$
\frac{\partial \phi}{\partial t} +
\frac{\partial W}{\partial x} =
J \left( -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
$$
Подставляя вместо $\frac{\partial \phi}{\partial t}$ правую часть уравнения неразрывности для породы:
$$
\frac{\partial W}{\partial x} =
J \left( -\frac{1}{\rho_1} -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
$$

Итого имеем 7 неизвестных: $\phi, S, c_2, c_3, c_5, J, W$ ;
\
4 параметра реакции $k_a, m, S_s, c_{\text{eq}}$
\
и 7 уравнений:

$$
\begin{gathered}
J = k_a S (1 - \phi) \frac{M_1}{2} S_s \rho_2^m \frac{(c_2 - c_{\text{eq}})^m}{M_2^m}
\\
\frac{\partial W}{\partial x} =
J \left( -\frac{1}{\rho_1} -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
\\
\frac{\partial \phi}{\partial t} = \frac{1}{\rho_1} J
\\
\frac{\partial \phi S}{\partial t} +
\frac{\partial Wb}{\partial x} =
J \left( -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
\\
\frac{\partial}{\partial t} \left[ \phi S c_2 \right] +
\frac{\partial}{\partial x} \left[  c_2 Wb \right] =
- \frac{\chi_2}{\rho_2} J
\\
\frac{\partial}{\partial t} \left[ \phi S c_3 \right] +
\frac{\partial}{\partial x} \left[ c_3 Wb \right] =
\frac{\chi_3}{\rho_3} J
\\
\frac{\partial}{\partial t} \left[ \phi S c_5 \right] +
\frac{\partial}{\partial x} \left[ c_5 Wb \right] =
\frac{\chi_5}{\rho_5} J
\end{gathered}
$$

<hr>

### Дискретизация

Дискретизируем неявно разностями против потока полученную систему:

$$
\begin{gathered}
J_i^n = k_a S_i^n (1 - \phi_i^n) \frac{M_1}{2} S_s \rho_2^m \frac{({c_2}_i^n - c_{\text{eq}})^m}{M_2^m}
\\
\frac{W_i^n - W_{i-1}^n}{h} =
J_i^n \left( -\frac{1}{\rho_1} -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
\\
\frac{\phi_i^{n+1} - \phi_i^n}{\tau} = \frac{1}{\rho_1} J_i^n
\\
\frac{\phi_i^{n+1} S_i^{n+1} - \phi_i^n S_i^n}{\tau} +
\frac{W_i^n b_i^n - W_{i-1}^n b_{i-1}^n}{h} =
J_i^n \left( -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
\\
\frac{\phi_i^{n+1} S_i^{n+1} {c_2}_i^{n+1} - \phi_i^n S_i^n {c_2}_i^n}{\tau} +
\frac{ {c_2}_i^n W_i^n b_i^n - {c_2}_{i-1}^n W_{i-1}^n b_{i-1}^n  }{h} =
- \frac{\chi_2}{\rho_2} J_i^n
\\
\frac{\phi_i^{n+1} S_i^{n+1} {c_3}_i^{n+1} - \phi_i^n S_i^n {c_3}_i^n}{\tau} +
\frac{ {c_3}_i^n W_i^n b_i^n - {c_3}_{i-1}^n W_{i-1}^n b_{i-1}^n  }{h} =
\frac{\chi_3}{\rho_3} J_i^n
\\
\frac{\phi_i^{n+1} S_i^{n+1} {c_5}_i^{n+1} - \phi_i^n S_i^n {c_5}_i^n}{\tau} +
\frac{ {c_5}_i^n W_i^n b_i^n - {c_5}_{i-1}^n W_{i-1}^n b_{i-1}^n  }{h} =
\frac{\chi_5}{\rho_5} J_i^n
\end{gathered}
$$

Алгоритм расчета:

1. Расчет скорости реакции

$$
J_i^n = k_a S_i^n (1 - \phi_i^n) \frac{M_1}{2} S_s \rho_2^m \frac{({c_2}_i^n - c_{\text{eq}})^m}{M_2^m}
$$

2. 
	- Расчет скорости фильтрации

	$$
	W_i^n = W_{i-1}^n +
	h J_i^n \left( -\frac{1}{\rho_1} -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
	$$

	- Расчет изменения пористости

	$$
	\phi_i^{n+1} = \phi_i^n + \frac{\tau}{\rho_1} J_i^n
	$$

2. Расчет водонасыщенности

$$
S_i^{n+1}
 = \frac{1}{\phi_i^{n+1}} \left\{ \phi_i^n S_i^n +
\tau \left[ J_i^n \left( -\frac{\chi_2}{\rho_2} +\frac{\chi_3}{\rho_3} + \frac{\chi_4}{\rho_4} + \frac{\chi_5}{\rho_5} \right)
- \frac{W_i^n b_i^n - W_{i-1}^n b_{i-1}^n}{h} \right] \right\}
$$

4. Расчет концентраций
   
$$
\begin{gathered}
{c_2}_i^{n+1} 
= \frac{1}{\phi_i^{n+1} S_i^{n+1}}
\left[ \phi_i^n S_i^n {c_2}_i^n - \tau \left(\frac{\chi_2}{\rho_2} J_i^n + \frac{ {c_2}_i^n W_i^n b_i^n - {c_2}_{i-1}^n W_{i-1}^n b_{i-1}^n }{h} \right) \right]
\\

{c_3}_i^{n+1} 
= \frac{1}{\phi_i^{n+1} S_i^{n+1}}
\left[ \phi_i^n S_i^n {c_3}_i^n + \tau \left( \frac{\chi_3}{\rho_3} J_i^n - \frac{ {c_3}_i^n W_i^n b_i^n - {c_3}_{i-1}^n W_{i-1}^n b_{i-1}^n }{h} \right) \right]
\\

{c_5}_i^{n+1} 
= \frac{1}{\phi_i^{n+1} S_i^{n+1}}
\left[ \phi_i^n S_i^n {c_5}_i^n + \tau \left(\frac{\chi_5}{\rho_5} J_i^n - \frac{ {c_5}_i^n W_i^n b_i^n - {c_5}_{i-1}^n W_{i-1}^n b_{i-1}^n }{h} \right) \right]
\end{gathered}
$$

#### Дополнительный ограничитель на шаг по времени


$$
\begin{gathered}
\phi_i^{n+1} = \phi_i^n + \frac{\tau}{\rho_1} J_i^n
\\
\phi_i^{n+1} ≤ 1
\\
\tau ≤ \frac{\rho_1}{J_i^n}(1 - \phi^n_i)
\\
\tau ≤ \frac{2 \rho_1 M_2^m}{k_a M_1 S_s \rho_2^m ({c_2}_{\text{input}} - c_{\text{eq}})^m}
\end{gathered}
$$

### Импорт библиотек

```python
%load_ext autoreload
%autoreload 2

import sys
if ".." not in sys.path:
	sys.path.insert(0, "..")

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numba

from src.BuckleyLeverett import *
from src.ExplicitAcidSolver import ExplicitAcidSolver
```

### Входные параметры

```python
EPS = 0.5 # параметр схемной вязкости [0, 1), чем больше - тем сильнее размываются разрывы, чем меньше - тем сильнее проявляются нефизичные осцилляции
```

```python
# Кол-во узлов сетки
num_of_grid_points = 100

# Геометрия
width = 0.1
height = 0.1
length = 0.5

# Хар-ки породы
# initial_permeability = 1 * 10**(-12)
initial_porosity = 0.2

rho1 = 2740
M1 = 0.100088

# Хар-ки флюидов
viscosity_water = 1 * 10**(-3)
viscosity_oil = 4 * 10**(-3)
M2 = 0.036461
rho2 = 1150
rho3 = 2510
rho4 = 1000
rho5 = 1167

# Расход
flow = 100 * 10**(-7)

# Насыщенность
saturation_left = 1 - S_ORW
saturation_initial = S_WIR
conc_left = 0.3

# Параметры реакции
k_a = 10**(-6)
m = 1
S_s = 30000
c_eq = 0
```

```python
s_right = plot_buckley_leverett(viscosity_water, viscosity_oil, return_Sfront=True, return_dfs=True)[0]
```

### Инициализация

```python
velocity = flow / (height * width)

# координатная сетка
x, h = np.linspace(0, length, num_of_grid_points, endpoint=True, retstep=True)

# оценка времени расчета
total_time = length * s_right / b(s_right, viscosity_water, viscosity_oil) / velocity * initial_porosity * 0.9
# total_time = 10

# временная сетка
tau_W = initial_porosity * h / (2 * velocity) * (1 - EPS)
tau_J = 2 * rho1 * M2**m / (k_a * M1 * S_s * rho2**m * (conc_left - c_eq)**m) / 50
tau = min(tau_W, tau_J)
# if tau_J < tau_W:
# 	print("Ограничитель по скорости реакции")
# else:
# 	print("Ограничитель по скорости переноса")
# t = np.arange(0, total_time + tau, tau)
# num_of_time_points = t.shape[0]
# print(f"{num_of_time_points} временных итераций")
num_of_time_points = 10

# Начальные условия
initial_porosity_field = np.zeros(num_of_grid_points) + initial_porosity
initial_saturation_field = np.zeros(num_of_grid_points) + saturation_initial + np.sin(x*100) / 2 + 0.5

# Граничные условия
# вектор скорости фильтрации
velocity_left = np.zeros(num_of_time_points) + velocity
# вектор водонасыщенности
water_saturation_left = np.zeros(num_of_time_points) + saturation_left
# вектор концентрации кислоты
acid_conc_left = np.zeros(num_of_time_points) + conc_left
```

### Вычисления

```python
# Инициализация модели
model = ExplicitAcidSolver(
    num_of_grid_points=num_of_grid_points,
    grid_step=h,

    initial_porosity_field=initial_porosity_field,
    initial_saturation_field=initial_saturation_field,

    k_a=k_a,
    m=m,
    S_s=S_s,
    c_eq=c_eq,
    water_viscosity=viscosity_water,
    oil_viscosity=viscosity_oil,
    molar_weight_of_carbonate=M1,
    molar_weight_of_acid=M2,
    carbonate_density=rho1,
    acid_density=rho2,
    salt_density=rho3,
    water_density=rho4,
    co2_density=rho5
)
```

```python
initial_saturation_field
```

```python
plt.plot(x, model.water_saturation_field)
plt.show()
```

### Визуализация

```python
solution = model.predict(
    num_of_time_steps=num_of_time_points,
    time_step=tau,
    water_saturation_left=water_saturation_left,
    acid_conc_left=acid_conc_left,
    fluid_velocity_left=velocity_left/2
)

fig, ax = plt.subplots(figsize=(10, 4), layout='constrained', dpi=300)

ax.plot(x, model.water_saturation_field, label="Saturation", marker="x", alpha=0.5)
ax.plot(x, model.porosity_field, label="Porosity", marker="x", alpha=0.5)
ax.plot(x, model.acid_concentration_field, label="HCl concentration", marker="x", alpha=0.5)
ax.plot(x, model.salt_concentration_field, label="CaCl${}_2$ concentration", marker="x", alpha=0.5)
ax.plot(x, model.co2_concentration_field, label="CO${}_2$ concentration", marker="x", alpha=0.5)

plt.title("Динамика поля насыщенности, концентрации и пористости")

textstr = '\n'.join((
	    r'$t=%.1f$ %s' % (model.time, "с"),
	    r'$Q = %.1f м^3/c$' % (flow),
	    r'$C_{\text{input}}=%.0f$ %s' % (model.acid_concentration_field[0]*100, "%"),
	    f'$k_a = {k_a}$',
	    f'$m = {m}$'
	    ))

ax.text(0.5, 0.95, textstr,
	horizontalalignment = 'center',
	verticalalignment = 'top',
	bbox={'facecolor': 'yellow', 'alpha': 0.2, 'pad': 5},
	transform=ax.transAxes)

ax.set_xlabel("x")
ax.set_ylim([-0.05, 1.05])
ax.set_xlim([-0.01 * length, length * 1.01])
# ax.set_xlim([9.7, 10.2])
plt.legend(fontsize=10)
plt.show()
```
